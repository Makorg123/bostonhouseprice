```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('boston.csv')
```

```python
dataset.head()
```

```python
dataset = dataset.rename(columns={'MEDV':'Price'})
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
# Summurizing the Stats of data

dataset.describe()
```

```python
# Checking the Missing values

dataset.isnull().sum()*100/len(dataset)
```

### EDA 

```python
plt.scatter(dataset['RM'],dataset['Price'])
plt.xlabel('RM')
plt.ylabel('Price')
plt.show()
```

```python
sns.regplot(x='RM',y='Price',data=dataset)
```

```python
sns.regplot(x='LSTAT',y='Price',data=dataset)
```

```python
# Correlatin
plt.figure(figsize=(14,7))
sns.heatmap(dataset.corr(),annot=True)
```

```python
# Splitting into Independent and Dependent features

X = dataset.iloc[:,:-1]
y = dataset['Price']
```

```python
X
```

```python
y
```

### Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
y_test
```

### Standardized the dataset

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test  = scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

## Model training

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
regmodel = reg.fit(X_train,y_train)
```

```python
regmodel.score(X_train,y_train)
```

```python
# Coefficient

print(reg.coef_)
```

```python
print(reg.intercept_)
```

```python
# Prediction with test data

reg_test_pred = regmodel.predict(X_test)
```

```python
reg_test_pred
```

```python
# Plot a scatter plot for prediction
plt.scatter(y_test,reg_test_pred)
```

```python
# Residual

residual = y_test - reg_test_pred
```

```python
residual
```

```python
# Plotting the residual
sns.displot(residual,kind='kde')
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
print(mean_absolute_error(y_test,reg_test_pred))
```

```python
print(mean_squared_error(y_test,reg_test_pred))
```

```python
# RMSE
print(np.sqrt(mean_squared_error(y_test,reg_test_pred)))
```

```python
# R square and Adjusted R square

# Formula

# R^2 = 1-SSR/SST

# R^2 = coefficient of determination SSR = Sum of Square of Residual, SST = Total Sum of Squares
```

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(y_test,reg_test_pred)
```

```python
score
```

```python
# Adjusted R Square

# Adjusted R2 = 1-[(1-R2)*(n-1)/(n-k-1)]

# where:
# R2: The R2 of the model 
# n: The number of observations 
# K: The number of predictor variables
```

```python
1-(1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
